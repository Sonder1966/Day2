```python
import numpy as np
import IPython.display as ipd

# Set constants
sample_rate = 44100  # Sample rate in Hz
duration = 5.0       # Duration in seconds
frequency = 440.0    # Frequency in Hz
harmonic_order = 20  # Order of harmonics
harmonic_amplitude = 0.1  # Amplitude of harmonics

# Generate time array
t = np.arange(int(sample_rate * duration)) / sample_rate

# Generate sine wave
input_signal = np.sin(2 * np.pi * frequency * t)

# Process the signal
output_signal = np.zeros_like(input_signal)
for z in range(1, harmonic_order+1):
    output_signal += harmonic_amplitude * z * np.power(input_signal, z+1)

# Normalize the signal
output_signal = output_signal / np.max(np.abs(output_signal))

# Convert to 16-bit PCM audio
audio = (output_signal * 32767).astype(np.int16)

# Play the sound
ipd.Audio(audio, rate=sample_rate)
```

```python
import numpy as np
import IPython.display as ipd

# Set constants
sample_rate = 44100  # Sample rate in Hz
duration = 5.0       # Duration in seconds
frequency = 440.0    # Frequency in Hz

# Generate time array
t = np.arange(int(sample_rate * duration)) / sample_rate

# Generate sine wave
signal = np.sin(2 * np.pi * frequency * t)

# Convert to 16-bit PCM audio
audio = (signal * 32767).astype(np.int16)

# Play the sound
ipd.Audio(audio, rate=sample_rate)
```
